```python
import torch.nn as nn

class FixedBoxConv3D(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.conv = nn.Conv3d(
            in_channels=channels,
            out_channels=channels,
            kernel_size=3,
            padding=1,
            groups=channels,
            bias=False
        )
        
        # initialize weights to 1
        with torch.no_grad():
            self.conv.weight.fill_(1.0 / 27.0)

        # freeze weights
        self.conv.weight.requires_grad = False

    def forward(self, x):
        # (B, C, H, W, D) -> (B, C, D, H, W)
        x = x.permute(0, 1, 4, 2, 3)
        x = self.conv(x)
        x = x.permute(0, 1, 3, 4, 2)
        return x
```

```python
import torch
import torch.nn.functional as F

def fixed_box_conv3d(x):
    # x: (B, C, H, W, D)

    # Step 1: rearrange to PyTorch format
    x = x.permute(0, 1, 4, 2, 3)  # (B, C, D, H, W)

    B, C, D, H, W = x.shape

    # Step 2: create fixed kernel (no grad)
    kernel = torch.ones((C, 1, 3, 3, 3), device=x.device)
    kernel = kernel / 27.0  # optional: normalize (mean filter)

    # Step 3: apply grouped convolution
    out = F.conv3d(
        x,
        kernel,
        bias=None,
        stride=1,
        padding=1,   # keeps same size
        groups=C     # critical: per-channel convolution
    )

    # Step 4: permute back
    out = out.permute(0, 1, 3, 4, 2)  # (B, C, H, W, D)

    return out
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class LocalGAT3D(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.proj = nn.Conv3d(in_channels, out_channels, 1)
        self.attn = nn.Linear(out_channels * 2, 1)

    def forward(self, x):
        h = self.proj(x)
        # extract neighbors (3x3x3)
        neighbors = self.extract_3d_patches(h)  # shape: B, C, 27, H, W, D
        print("Neighbour shape:", neighbors.shape)
        h_center = h.unsqueeze(2).expand_as(neighbors)
        print("Center shape:", h_center.shape)
        concat = torch.cat([h_center, neighbors], dim=1).permute(0,3,4,5,2,1)  # (B, H, W, D, 27, 2C)
        print("Concat shape:", concat.shape)
        e = self.attn(concat).permute(0, 5, 4, 1, 2, 3)
        print("Attention shape:", e.shape)
        alpha = torch.softmax(e, dim=2)
        print("Alpha shape:", alpha.shape)
        out = (alpha * neighbors).sum(dim=2)
        return out

    def extract_3d_patches(self, x, kernel_size=3, padding=1):
        """
        x: (B, C, D, H, W)
        returns: (B, C, K^3, D, H, W)
        """
        B, C, D, H, W= x.shape
        K = kernel_size

        # Pad input
        x = F.pad(x, (padding, padding, padding, padding, padding, padding))

        # New spatial dims after padding
        Dp, Hp, Wp = x.shape[2:]

        # Compute strides
        stride_b, stride_c, stride_d, stride_h, stride_w = x.stride()
        # Create strided view
        patches = x.as_strided(
            size=(B, C, D, H, W, K, K, K),
            stride=(
                stride_b,
                stride_c,
                stride_d,
                stride_h,
                stride_w,
                stride_d,
                stride_h,
                stride_w,
            ),
        )

        # reshape to (B, C, 27, D, H, W)
        patches = patches.contiguous().view(B, C, D, H, W, -1)
        patches = patches.permute(0, 1, 5, 2, 3, 4)

        return patches
    # def extract_3d_patches(self, x):
    #     B, C, H, W, D = x.shape

    #     # Create identity kernels (27 filters)
    #     weight = torch.zeros(27, 1, 3, 3, 3, device=x.device)

    #     idx = 0
    #     for i in range(3):
    #         for j in range(3):
    #             for k in range(3):
    #                 weight[idx, 0, i, j, k] = 1
    #                 idx += 1

    #     # Apply grouped conv
    #     x = x.view(B * C, 1, H, W, D)
    #     patches = torch.conv3d(x, weight, padding=1)

    #     patches = patches.view(B, C, 27, H, W, D)

    #     return patches
```
